```python
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None, names=['sepal length','sepal width','petal length','petal width','class'])
iris.head()
```

```python
iris.groupby('class')
```

```python
iris.groupby('class').mean()
```

```python
iris_groups = iris.groupby('class')
iris_groups.groups
```

```python
iris_groups.get_group('Iris-setosa').head(10)
```

```python
iris_groups.mean()
```

```python
iris_groups.median()
```

```python
def longest_petal(g):
    return g['petal length'].idxmax()

iris_groups.apply(longest_petal)
```

```python
def longest_petal_length(g):
    return g.loc[g['petal length'].idxmax()]

iris_groups.apply(longest_petal_length)
```

```python
iris.groupby('class')['petal length'].mean()
```

```python
iris.groupby('class')['petal length'].aggregate(['min', 'mean', 'max'])
```

```python
iris.groupby('class')[['petal length', 'sepal length']].aggregate(['min', 'mean', 'max'])
```

```python
TObike = pd.read_excel('2016_Bike_Share_Toronto_Ridership_Q4.xlsx')
TObike.head(10)
```

```python
TObike.info()
```

```python
TObike.trip_start_time = pd.to_datetime(TObike.trip_start_time)
TObike.trip_start_time
```

```python
TObike['weekday'] = TObike.trip_start_time.dt.day_name()
TObike.head()
```

```python
TObike['weekday'].value_counts().idxmax()
```

```python
TObike['trip_duration_seconds'].groupby(TObike['weekday']).aggregate(sum).sort_values(ascending=False).idxmax()
```
