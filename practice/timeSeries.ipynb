```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
longTimeRange = pd.date_range("1/1/2015", periods=100, freq="H")
aTimeSeries = pd.Series(np.random.randn(len(longTimeRange)), index=longTimeRange)
aTimeSeries.head(10)
```

```python
# Time plot
aTimeplot = aTimeSeries.plot(style="-o", title="A Time Plot (Random Data)")
aTimeplot.set_ylabel("Values (No Units)")
aTimeplot.set_xlabel("Time (Days)")
aTimeplot.plot()
```

```python
aTimeframe = aTimeSeries.to_frame(name="value")
aTimeframe.head(10)
```

```python
# convert series to data frame
# `pivot_table` is used to make each season its own column. And, plot.
# Recall the `melt` function. This is a complementary function for reshaping data.
aSeasonalFrame = aTimeframe.pivot_table(values="value", aggfunc='sum', index=aTimeframe.index.hour, columns=aTimeframe.index.day)
aSeasonalPlot = aSeasonalFrame.plot(style="-o", title="A Seasonal Plot (Same Random Data as before)")
aSeasonalPlot.set_ylabel("Values (No Units)")
aSeasonalPlot.set_xlabel("Time (Hours)")
aSeasonalPlot.plot()
aSeasonalFrame.columns.name = "Seasons (Day #)"
aSeasonalFrame
```

```python
# lag plot
aLagPlot = pd.plotting.lag_plot(series=aTimeSeries, lag=1)
aLagPlot.set_ylabel("Values (No Units)")
aLagPlot.set_xlabel("Time (Days)")
aLagPlot.plot()
```

## Time series, Time plot vs seasonal plot

```python
timeRange = pd.date_range(start='2015-01-01', periods=100, freq='H')
timeRange
```

```python
timeSeries = pd.Series(np.random.randn(len(timeRange)), index=timeRange)
timeSeries.head(10)
```

```python
timePlot = timeSeries.plot(style='-o', title="A Time Plot (Random Data)")
timePlot.set_ylabel("Values (No Units)")
timePlot.set_xlabel("Time (Days)")
timePlot.plot()
```

```python
timeFrame = timeSeries.to_frame(name="value")
seasonalFrame = timeFrame.pivot_table(values="value", aggfunc='sum', index=timeFrame.index.hour, columns=timeFrame.index.day)
seasonalFrame.columns.name = "Seasons (Day #)"
seasonalPlot = seasonalFrame.plot(style="-o", title="A Seasonal Plot (Same Random Data as before)")
seasonalPlot.set_ylabel("Values (No Units)")
seasonalPlot.set_xlabel("Time (Hours)")
seasonalPlot.plot()
```

## lag plots


```python
aLagPlot = pd.plotting.lag_plot(series=timeSeries, lag=1)
aLagPlot.set_ylabel("Values (No Units)")
aLagPlot.set_xlabel("Time (Days)")
aLagPlot.plot() 
```

## Drift

```python
h = 1
numerator = timeSeries.last(offset="H")[0] - timeSeries.first(offset="H")[0]
denominator = np.size(timeSeries) - 1

trendSlope = numerator / denominator
print(trendSlope)

(timeSeries + h * trendSlope).shift(h).last(offset="H")

```

```python
# Get GDP data
gdpUrl = "https://raw.githubusercontent.com/datasets/gdp/master/data/gdp.csv"
gdpDf = pd.read_csv(gdpUrl)
canadaGdpDf = gdpDf[gdpDf["Country Name"] == "Canada"]
canadaGdpDf
```

```python
# get population data
popUrl = "https://raw.githubusercontent.com/datasets/population/master/data/population.csv"
popCsv = pd.read_csv(popUrl)
canadaPopDf = popCsv[popCsv["Country Name"] == "Canada"]
canadaPopDf
```

```python
# calculate gdp per person
canadaGDPPerPersonDF = canadaGdpDf.merge(canadaPopDf, on="Year")
canadaGDPPerPersonDF["GDP per person"] = canadaGDPPerPersonDF["Value_x"] / canadaGDPPerPersonDF["Value_y"]
canadaGDPPerPersonDF["GDP (USD)"] = canadaGDPPerPersonDF["Value_x"]
canadaGDPPerPersonDF["Population"] = canadaGDPPerPersonDF["Value_y"]
canadaGDPPerPersonDF =canadaGDPPerPersonDF.drop(columns=["GDP per capita"])
canadaGDPPerPersonDF
```

```python
## plot

## GDP
canadaGDPPlot = canadaGDPPerPersonDF.plot(x="Year", y="GDP (USD)", style="-o", title="GDP per Person (Canada)")
canadaGDPPlot.set_ylabel("GDP (USD)")
canadaGDPPlot.set_xlabel("Year")
canadaGDPPlot.plot()
```

```python
# Population
canadaPopPlot = canadaGDPPerPersonDF.plot(x="Year", y="Population", style="-o", title="Population (Canada)")
canadaPopPlot.set_ylabel("Population")
canadaPopPlot.set_xlabel("Year")
canadaPopPlot.plot()
```

```python
# GDP per person
canadaGDPPerPersonPlot = canadaGDPPerPersonDF.plot(x="Year", y="GDP per person", style="-o", title="GDP per Person (Canada)")
canadaGDPPerPersonPlot.set_ylabel("GDP per Person")
canadaGDPPerPersonPlot.set_xlabel("Year")
canadaGDPPerPersonPlot.plot()
```
