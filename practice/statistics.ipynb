```python
a = np.linspace(-np.pi, np.pi, 200, endpoint=True)
cosine, sine = np.cos(a), np.sin(a)
plt.plot(a, cosine, a, sine)
plt.figure(figsize=(10,6))
plt.plot(a, cosine, linestyle='dashed')     # plotting cosine function, red dashed line
plt.plot(a, sine, linestyle='dotted')       # plotting sine function, blue dotted line
```

```python
plt.plot(a, cosine, label = 'cosine')                 # adding label parameter
plt.plot(a, sine, label = 'sine', color = 'green')    # adding the color parameter and label

plt.xlabel('Angle (radians)')
plt.ylabel('Amplitude')
plt.title('Cosine and Sine')
plt.legend()
```

```python
plt.plot(a, cosine, label = 'cosine')                 # adding label parameter
plt.plot(a, sine, label = 'sine', color = 'green')    # adding the color parameter and label

plt.xlabel('Angle (radians)')
plt.ylabel('Amplitude')
plt.title('Cosine and Sine')
plt.legend(loc='upper left')
```

```python
plt.plot(a, cosine, label = 'cos(a)')                 # adding label parameter
plt.plot(a, sine, label = 'sin(a)', color = 'green')    # adding the color parameter and label

plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Cosine and Sine')
plt.legend(loc='upper left')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
       [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

plt.show()
```
