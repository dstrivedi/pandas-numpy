```python
import pandas as pd
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url, sep="\t")
chipo
```

```python
# What is the number of observations in the dataset?
print("What is the number of observations in the dataset?")
print(chipo.shape[0])
```

```python
#  What is the number of columns in the dataset?
print(" What is the number of columns in the dataset?")
print(chipo.shape[1])
```

```python
# Print the name of all the columns.
print("Print the name of all the columns.")
print(chipo.columns)
```

```python
# How is the dataset indexed?
print("How is the dataset indexed?")
print(chipo.index)
```

```python
# Which was the most-ordered item?
items = chipo.groupby("item_name")
items = items.sum()
items = items.sort_values(['quantity'], ascending=False)
print("Which was the most-ordered item?")
print(items.head(1))
```

```python
# What was the most ordered item in the choice_description column?
print("What was the most ordered item in the choice_description column?")
cd = chipo.groupby('choice_description').sum()
cd = cd.sort_values(['quantity'], ascending=False)
cd.head(1)
```

```python
# How many items were orderd in total?
print("How many items were orderd in total?")
total_items_order = chipo.quantity.sum()
total_items_order
```

```python
# Turn the item price into a float
print("Turn the item price into a float")
print(chipo.item_price.dtype)
# chipo.item_price.astype(dtype='f') error
# Remove $ and spaces, then convert to float
chipo['item_price'] = chipo.item_price.str.replace("$", "", regex=False).str.strip().astype(float)
print(chipo.item_price.dtype)
```

```python
# How many orders were made in the period?
print("How many orders were made in the period?")
total_orders = chipo.order_id.value_counts().count()
print(total_orders)
```

```python
# What is the average revenue amount per order?
print("What is the average revenue amount per order?")
chipo['revenue'] = chipo['quantity'] * chipo['item_price']
chipo.groupby('order_id')['revenue'].sum().mean()
```

```python
# how many different items are sold?
print("how many different items are sold?")
print(chipo['item_name'].unique().size)
# OR
print(chipo['item_name'].value_counts().count())
```

```python
# How many products cost more than $10.00?
print("How many products cost more than $10.00?")

# With duplicates
print("This is with duplicates")
cond = chipo[chipo['item_price'] > 10]
print(cond.shape[0])# using shape[0] is better than size here, because size counts all elements (rows Ã— columns)
# print(cond)

# Without duplicates
print("Without duplicates")
chipo_filtered = chipo.drop_duplicates(['item_name', 'quantity', 'choice_description'])
chipo_filtered = chipo_filtered[chipo_filtered['item_price'] > 10]
print(chipo_filtered.shape[0])

# After cleaning data
print("After cleaning data and remove duplicates")
chipo.dtypes['item_price']
filtered_items = (
    chipo
    .dropna()
    # .drop_duplicates(['item_name', 'quantity', 'choice_description'])
    .query('item_price > 10')
    [['item_name', 'item_price']]
    .sort_values('item_price')
)
filtered_items
```

```python
# What is the price of each item?
print("What is the price of each item?")
item = (
    chipo
    .query("quantity == 1")
    .drop_duplicates(["item_name"])
    [["item_name", "item_price"]]
)
item
```

```python
# Sort by the name of the item
print("Sort by the name of the item")
chipo.sort_values("item_name")
```

```python
# What was the quantity of the most expensive item ordered?
most_expensive_item = chipo.sort_values("item_price", ascending=False).iloc[0]['quantity']
# sort_data.head(1)['quantity']
most_expensive_item
```

```python
# How many times was a Veggie Salad Bowl ordered?
veggie_salad_bowl = (
    chipo
    .query("item_name == 'Veggie Salad Bowl'")
)
print(veggie_salad_bowl.value_counts().count())

# OR 

veggie_salad = chipo[chipo.item_name == "Veggie Salad Bowl"]
print(len(veggie_salad))
```

```python
# How many times did someone order more than one Canned Soda?
print("How many times did someone order more than one Canned Soda?")
canned_soda = chipo[(chipo.item_name == "Canned Soda") & (chipo.quantity > 1)]
len(canned_soda)
```
