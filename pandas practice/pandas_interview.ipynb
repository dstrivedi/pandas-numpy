5.  How do you handle null or missing values in pandas?

- dropna() – the function removes the missing rows or columns from the DataFrame.

- fillna() – fill nulls with a specific value using this function.

- interpolate() – this method fills the missing values with computed interpolation values. The interpolation technique can be linear, polynomial, spline, time, etc.,

different b/w fillna() and interpolate() –

- fillna() fills the missing values with the given constant. Plus, you can give forward-filling or backward-filling inputs to its ‘method’ parameter.

- interpolate() By default, this function fills the missing or NaN values with the linear interpolated values. However, you can customize the interpolation technique to polynomial, time, index, spline, etc., using its ‘method’ parameter.

The interpolation method is highly suitable for time series data, whereas fillna is a more generic approach.

7. Given a DataFrame with missing values, fill numeric columns with their mean and categorical columns with "Unknown".

```python
df = pd.DataFrame({
    "A": [1, 2, None, 4],
    "B": ["X", None, "Y", "Z"]
})

df['B'] = df['B'].fillna('Unknown')

mean = df['A'].mean()

df['A'] = df['A'].fillna(mean)

df
```

```python
# using function

df = pd.DataFrame({
    "A": [1, 2, None, 4],
    "B": ["X", None, "Y", "Z"]
})

def fill_na(x):
    for col in x.columns:
        if x[col].dtype == 'object':
            x[col].fillna('Unknown', inplace = True)
        else:
            x[col].fillna(x[col].mean(), inplace = True)
            
fill_na(df)
print(df)
```

13. Using method chaining, filter products with price > 100, group by category, and get the top 2 most expensive in each category.

```python
products = pd.DataFrame({
    "Category": ["Electronics", "Electronics", "Clothing", "Clothing"],
    "Product": ["TV", "Laptop", "Shirt", "Pants"],
    "Price": [1200, 1500, 30, 45]
})

products
```

```python
filtered_products = (
        products.where(products['Price'] > 100)
        .dropna()
        .sort_values(['Price'],ascending=False)
        .groupby(products['Category'])
        .head(2)
    )
filtered_products
```

```python
result = (
    products[products['Price'] > 100]
    .sort_values(['Category', 'Price'], ascending=[True, False])
    .groupby('Category')
    .head(2)
)
result
```

21.  Analyze stock prices for multiple companies across time.

```python
dates = pd.date_range('2023-01-01', periods=5)
stocks = ['AAPL', 'GOOG']
index = pd.MultiIndex.from_product([stocks, dates], 
                                 names=['Stock', 'Date'])

prices = pd.DataFrame({
    'Open': np.random.uniform(100, 200, 10),
    'Close': np.random.uniform(100, 200, 10)
}, index=index)
prices
```

- Calculate daily returns

```python
returns = prices.groupby(level='Stock').apply(
    lambda x: x['Close'].pct_change())
returns
```

```python
prices.groupby(level='Stock').agg('mean')
```
