```python
cust = pd.read_csv('customers-100.csv')
cust
```

```python
cust.info()
```

```python
cust['Subscription Date'] = pd.to_datetime(cust['Subscription Date'])
cust
```

```python
# cust['Date'] = cust['Subscription Date'].dt.date
cust['Year'] = cust['Subscription Date'].dt.year
cust['Day'] = cust['Subscription Date'].dt.day
cust['Day_name'] = cust['Subscription Date'].dt.day_name()
cust['Month'] = cust['Subscription Date'].dt.month_name()
cust['Month_no'] = cust['Subscription Date'].dt.month
cust
```

```python
df = pd.DataFrame(cust['Subscription Date'])
df['Year'] = df['Subscription Date'].dt.year
df['Day'] = df['Subscription Date'].dt.day
df['Day_name'] = df['Subscription Date'].dt.day_name()
df['Month'] = df['Subscription Date'].dt.month_name()
df['Month_no'] = df['Subscription Date'].dt.month
df['Date_dow'] = df['Subscription Date'].dt.dayofweek
df['Date week'] = df['Subscription Date'].dt.isocalendar().week
df
```

```python
# Extracting quarter and semester
df['Quarter'] = df['Subscription Date'].dt.quarter
df
```

```python
# # Checking if the date is a weekend
df['Date_is_weekend'] = np.where(df['Day_name'].isin(['Sunday', 'Saturday']), 1 , 0)
df
```
